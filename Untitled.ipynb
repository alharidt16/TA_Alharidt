```python
import numpy as np

def k_medians(X, k):
    # randomly initialize k medians
    medians = X[np.random.choice(X.shape[0], k, replace=False)]
    while True:
        # assign each point to the closest median
        clusters = [[] for _ in range(k)]
        for x in X:
            distances = [np.linalg.norm(x - m) for m in medians]
            closest_median_idx = np.argmin(distances)
            clusters[closest_median_idx].append(x)
        # recompute the medians
        new_medians = np.array([np.median(cluster, axis=0) for cluster in clusters])
        # check if the medians have changed
        if np.all(new_medians == medians):
            break
        else:
            medians = new_medians
    return clusters
```
